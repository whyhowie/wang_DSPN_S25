## 3. Linear regression (4 points)

Use the `lm` (linear model) function to determine the association between Flanker Task performance and total grey matter volume from the `d1` data frame.

$$
Y_{flanker} = \beta_0 + \beta_1 X_{GM volume}
$$

Show the results using the `summary` function, and report the mean coefficient values for $\beta_0$ \& $\beta_1$ (`coef` function) and their 95\% confidence intervals (`confint` function). Is grey matter volume significantly associated with Flanker Task performance?

```r
#INSERT CODE HERE
flanker_lm  <- lm(Flanker_Unadj ~ FS_Total_GM_Vol, d1)
summary(flanker_lm)

coef(flanker_lm)
confint(flanker_lm)
```

> *Write your response here.* 

> The mean coefficient values are: $\beta_0 \approx 90.256$ and $\beta_1 \approx 3.110\times10^{-5}$.
> The $95\%$ confidence interval for $\beta_0$ is $[84.289, 96.224]$; for $\beta_1$, it is $[2.244\times10^{-5}, 3.976\times10^{-5}]$

## 4. Plotting (2 points)

Use `ggplot` to plot the `FS_Total_GM_Vol` variable (x axis) against the `Flanker_Unadj` variable (y axis), as well as the regression line with confidence intervals on the regression line. Qualitatively describe what you see. 

```r
#INSERT CODE HERE
library(ggplot2)

ggplot(d1, aes(x = FS_Total_GM_Vol, y = Flanker_Unadj)) +
  geom_point() +                                # Scatter plot
  geom_abline(intercept = coef(flanker_lm)[1], slope = coef(flanker_lm)[2], color = "blue") +  # Regression line
  labs(title = "Flanker Task Performance",
       x = "Total Grey Matter Volume",
       y = "Flanker Task Performance")
```

> *Write your response here.* 

> The plot is indeed reflecting some positive correlation between the total grey matter volume and the Flanker task performance. However, the correlation appears very weak and regression coefficient ($\beta_1$) has a small value. We can only say that there is very little linear relationship between the two variables.
